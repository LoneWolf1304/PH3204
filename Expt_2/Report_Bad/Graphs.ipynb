```python
import pandas as pd
import matplotlib.pyplot as plt
V2 = pd.read_excel("../Data/Data.ods",sheet_name="VCE=2")
V3 = pd.read_excel("../Data/Data.ods",sheet_name="VCE=3")
I10 = pd.read_excel("../Data/Data.ods",sheet_name="IB=10")
I25 = pd.read_excel("../Data/Data.ods",sheet_name="IB=25")
I30 = pd.read_excel("../Data/Data.ods",sheet_name="IB=30")
I40 = pd.read_excel("../Data/Data.ods",sheet_name="IB=40")
```

```python
plt.plot(V2["V_BE (V)"], V2["I_B (micro)"],"ro-",label=r"V$_\text{CE}$ = 2 V")
plt.plot(V3["V_BE"], V3["I_B (micro)"],"bo-",label=r"V$_\text{CE}$ = 3 V")
plt.legend()
plt.grid()
plt.xlabel(r"$\mathrm{V_{BE}}$ (Volt)")
plt.ylabel(r"$\mathrm{I_{B}}$ (mA)")
plt.savefig("./Graphs/Input.svg")
```

```python
plt.plot(I25["V_CE (milli)"],I25["I_C (micro)"],"ro-",label=r"$\mathrm{I_B = 25 \ \mu A}$")
plt.plot(I30["V_CE (milli)"], I30["I_C (micro)"],"bo-",label=r"$\mathrm{I_B = 30 \ \mu A}$")
plt.plot(I40["V_CE (milli)"], I40["I_C (micro)"],"go-",label=r"$\mathrm{I_B = 40 \ \mu A}$")
plt.axvline(0.5,ls="--",color="black")
plt.grid()
plt.legend()
plt.xlabel(r"$\mathrm{V_{CE}}$ (Volt)")
plt.ylabel(r"$\mathrm{I_C \ (\mu A)}$ ")
plt.savefig("./Graphs/Output.svg")
```
