```julia
using Plots, CSV, DataFrames, LaTeXStrings
pgfplotsx()
Plots.PGFPlotsXBackend()
```

```julia
#input characteristics 
vce_2 = CSV.read("input2.csv", DataFrame)
vce_3 = CSV.read("input3.csv", DataFrame)

#output characteristics
ib_25 = CSV.read("output25.csv", DataFrame)
ib_30 = CSV.read("output30.csv", DataFrame)
ib_40 = CSV.read("output40.csv", DataFrame);
```

```julia
plot( [vce_2[!,2] vce_3[!,2]], [vce_2[!,3] vce_3[!,3]] , marker=[:circle :circle], lw=2, ms=3, c=["red" "black"], label=[L"\mathrm{VCE=3 \ V}" L"\mathrm{VCE=2 \ V}"] , msw=0.5, xticks=0:0.1:1, yticks=0:50:500, ylabel=L"\mathrm{I_B} \ (\mu A)", xlabel=L"\mathrm{V_{BE} \ (V)}", size=(600,400), grid=:true, gridwidth=3)
```

```julia
savefig("input_characteristics.tex")
```

```julia
ind = sortperm(ib_25[!,4])
plot(ib_25[!,4][ind] , ib_25[!,5][ind], marker=:circle, lw=2, ms=3, c="red", label=L"\mathrm{I_B}=25\ \mathrm{\mu A}", msw=0.5)
ind = sortperm(ib_30[!,4])
plot!(ib_30[!,4][ind] , ib_30[!,5][ind], marker=:circle, lw=2, ms=3, c="blue", label=L"\mathrm{I_B}=30\ \mathrm{\mu A}", msw=0.5)
ind = sortperm(ib_40[!,4])
plot!(ib_40[!,4][ind] , ib_40[!,5][ind], marker=:circle, lw=2, ms=3, c="green", label=L"\mathrm{I_B}=40\ \mathrm{\mu A}", msw=0.5)
plot!(xlabel =L"\mathrm{V_{CE}}\  \mathrm{(V)}", ylabel=L"\mathrm{I_C\  (mA) }", title="Output Characteristics of Transistor", grid=:true, gridwidth=3, xticks=0:0.1:1, yticks=0:1000:6500, size = (600, 400))
```
